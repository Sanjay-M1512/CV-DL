```python
import numpy as np

def perceptron(x, w, b):
    z = np.dot(x, w) + b
    print("Weighted sum:", z)
    print("Output:", 1 if z > 0 else 0)

x = np.array([1,2,3])
w = np.array([0.5, 0.3, 0.1])
b = 0.5

perceptron(x, w, b)
```

```python
import numpy as np

def perceptron(x, w, b):
    z = np.dot(x, w) + b
    print("Weighted sum:", z)
    print("Output:", 1 if z > 0 else 0)

x = np.array([1,1,1])
w1 = np.array([0.5, 0.3, 0.8])
w2 = np.array([0.5, 0.3, 0.1])
b = 5.0

perceptron(x, w1, b)
perceptron(x, w2, b)
```

```python
import numpy as np

pred=np.array([3.5, 2.1, 4.8])
actual=np.array([3.0, 2.5, 4.0])

actual = np.array([1, 0, 1, 1])
pred   = np.array([0.9, 0.2, 0.8, 0.7])

actual1 = np.array([1, 0, 1, 1])
pred1   = np.array([0.9, 0.2, 0.8, 0.7])

MSE = np.mean((pred - actual) ** 2)
Entropy = -np.mean(actual1 * np.log(pred1) + (1 - actual1) * np.log(1 - pred1))

print("Mean Squared Error:", round(MSE, 2))
print("Entropy:", round(Entropy, 2))
```
