### ***Day1***

```python
import numpy as np

x=np.array([2,3,1])
w=np.array([0.5,0.3,0.8])
b=0.5

z=np.dot(x,w)+b
print("Weighted sum:",z)

if z>0:
    print("Output: 1")
else:   
    print("Output: 0")
```

### ***Qn:1***

```python
import numpy as np

def perceptron(x, w, b):
    z = np.dot(x, w) + b
    print("Weighted sum:", z)
    print("Output:", 1 if z > 0 else 0)

x = np.array([1,2,3])
w = np.array([0.5, 0.3, 0.1])
b = 0.5

perceptron(x, w, b)
```
