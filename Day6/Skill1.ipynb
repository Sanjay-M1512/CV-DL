```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

data = np.load(r"C:\Users\HP\OneDrive\Desktop\CV&DL\student_micro_dataset2.npz")
mnist_x = data['mnist']   # shape: (50, 28, 28)
cifar_x = data['cifar']   # shape: (50, 32, 32, 3)

# Display 5 MNIST images
fig, axes = plt.subplots(1, 5, figsize=(12, 3))
fig.suptitle("MNIST Images (Grayscale)", fontsize=13)
for i in range(5):
    axes[i].imshow(mnist_x[i], cmap='gray')
    axes[i].set_title(f"Image {i+1}")
    axes[i].axis('off')
plt.tight_layout()
plt.show()

# Display 5 CIFAR images (RGB)
fig, axes = plt.subplots(1, 5, figsize=(12, 3))
fig.suptitle("CIFAR Images (RGB)", fontsize=13)
for i in range(5):
    axes[i].imshow(cifar_x[i])
    axes[i].set_title(f"Image {i+1}")
    axes[i].axis('off')
plt.tight_layout()
plt.show()

# Convert CIFAR images to grayscale and display
fig, axes = plt.subplots(1, 5, figsize=(12, 3))
fig.suptitle("CIFAR Images (Grayscale)", fontsize=13)
for i in range(5):
    gray = cv2.cvtColor(cifar_x[i], cv2.COLOR_RGB2GRAY)
    axes[i].imshow(gray, cmap='gray')
    axes[i].set_title(f"Image {i+1}")
    axes[i].axis('off')
plt.tight_layout()
plt.show()
```
