```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

data = np.load(r"C:\Users\HP\OneDrive\Desktop\CV&DL\student_micro_dataset2.npz")
mnist_x = data['mnist']  
cifar_x = data['cifar']  

fig, axes = plt.subplots(2, 5, figsize=(14, 5))
fig.suptitle("MNIST  |  Top: Original   Bottom: Canny Edges", fontsize=12)
for i in range(5):
    edges = cv2.Canny(mnist_x[i], 50, 150)
    axes[0, i].imshow(mnist_x[i], cmap='gray')
    axes[0, i].set_title(f"Orig {i+1}")
    axes[0, i].axis('off')
    axes[1, i].imshow(edges, cmap='gray')
    axes[1, i].set_title(f"Edges {i+1}")
    axes[1, i].axis('off')
plt.tight_layout()
plt.show()

fig, axes = plt.subplots(3, 5, figsize=(14, 8))
fig.suptitle("CIFAR-10  |  Top: Original RGB   Mid: Grayscale   Bottom: Canny Edges", fontsize=12)
for i in range(5):
    gray  = cv2.cvtColor(cifar_x[i], cv2.COLOR_RGB2GRAY)
    edges = cv2.Canny(gray, 50, 150)
    axes[0, i].imshow(cifar_x[i])
    axes[0, i].set_title(f"Orig {i+1}")
    axes[0, i].axis('off')
    axes[1, i].imshow(gray, cmap='gray')
    axes[1, i].set_title(f"Gray {i+1}")
    axes[1, i].axis('off')
    axes[2, i].imshow(edges, cmap='gray')
    axes[2, i].set_title(f"Edges {i+1}")
    axes[2, i].axis('off')
plt.tight_layout()
plt.show()
```
