```python
cifar_h, cifar_w = 32, 32
fashion_h, fashion_w = 28, 28

print(f"CIFAR-10   Original: {cifar_h}×{cifar_w}   → Resized: {cifar_h//2}×{cifar_w//2}")
print(f"Fashion-MNIST Original: {fashion_h}×{fashion_w} → Resized: {fashion_h//2}×{fashion_w//2}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

data = np.load(r"C:\Users\HP\OneDrive\Desktop\CV&DL\student_micro_dataset2.npz")
cifar_x   = data['cifar']    # (50, 32, 32, 3)
fashion_x = data['fashion']  # (50, 28, 28)

# ── CIFAR-10 ──────────────────────────────────────────────────────────────
fig, axes = plt.subplots(2, 5, figsize=(14, 5))
fig.suptitle("CIFAR-10  |  Top: Original (32×32)   Bottom: Resized (16×16)", fontsize=12)
for i in range(5):
    original = cifar_x[i]
    resized  = cv2.resize(original, (16, 16), interpolation=cv2.INTER_AREA)
    axes[0, i].imshow(original)
    axes[0, i].set_title(f"Orig {i+1}\n32×32")
    axes[0, i].axis('off')
    axes[1, i].imshow(resized)
    axes[1, i].set_title(f"Resized {i+1}\n16×16")
    axes[1, i].axis('off')
plt.tight_layout()
plt.show()

# ── Fashion-MNIST ─────────────────────────────────────────────────────────
fig, axes = plt.subplots(2, 5, figsize=(14, 5))
fig.suptitle("Fashion-MNIST  |  Top: Original (28×28)   Bottom: Resized (14×14)", fontsize=12)
for i in range(5):
    original = fashion_x[i]
    resized  = cv2.resize(original, (14, 14), interpolation=cv2.INTER_AREA)
    axes[0, i].imshow(original, cmap='gray')
    axes[0, i].set_title(f"Orig {i+1}\n28×28")
    axes[0, i].axis('off')
    axes[1, i].imshow(resized, cmap='gray')
    axes[1, i].set_title(f"Resized {i+1}\n14×14")
    axes[1, i].axis('off')
plt.tight_layout()
plt.show()
```
