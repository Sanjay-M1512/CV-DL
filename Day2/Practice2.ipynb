```python
import numpy as np
actual = 1
pred = 0.6

MSE = (pred - actual) ** 2
Entropy = - (actual * np.log(pred) + (1 - actual) * np.log(1 - pred))

print("Mean Squared Error:", round(MSE, 2))
print("Binary Cross Entropy Loss:", round(Entropy, 4))

w=0.8
n=0.1
dw=0.5
nn=0.5
g= w - n * dw
gn= w - nn * dw
print("Updated weight (n=0.1) :", round(g, 2))
print("Updated weight (nn=0.5):", round(gn, 2))
```
