```python
import math

y = 0
y_hat = 0.3
x = 1.5
w = 0.6
lr = 0.2

dL_dy_hat = (y_hat - y) / (y_hat * (1 - y_hat))
dy_hat_dw = x
dL_dw = dL_dy_hat * dy_hat_dw
w_new = w - lr * dL_dw

print(f'dL/dy_hat = {dL_dy_hat:.4f}')
print(f'dy_hat/dw = {dy_hat_dw}')
print(f'dL/dw = {dL_dw:.4f}')
print(f'Updated weight = {w_new:.4f}')
```
