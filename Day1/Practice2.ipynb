```python
import numpy as np

def sigmoid(z):
    return 1/(1+np.exp(-z))

def relu(z):
    return np.maximum(0, z)

def tanh(z):
    return np.tanh(z)

x=np.array([0.5,1.5])
w=np.array([-0.4,0.9])
b=0.2

z=np.dot(x,w)+b
output_sigmoid=sigmoid(z)
output_relu=relu(z)
output_tanh=tanh(z)

print("Linear Combination (z):", z)
print("Sigmoid Output:", output_sigmoid)
print("ReLU Output:", output_relu)
print("Tanh Output:", output_tanh)
```
