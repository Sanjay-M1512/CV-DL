```python
import numpy as np

def sigmoid(z): 
    return 1 / (1 + np.exp(-z))

def relu(z): 
    return np.maximum(0, z)

x = np.array([1.5, -2.0])

w1 = np.array([0.3, -0.8])
b1 = 0.2

w2 = 0.7
b2 = -0.1

z1 = np.dot(x, w1) + b1
a1 = relu(z1)
z2 = a1 * w2 + b2
output = sigmoid(z2)

print("Z1:", z1)
print("A1 (ReLU):", a1)
print("Z2:", z2)
print("Final Output (Sigmoid):", round(output, 4))
```
