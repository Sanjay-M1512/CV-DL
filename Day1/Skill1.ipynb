```python
import numpy as np

def sigmoid(z): 
    return 1 / (1 + np.exp(-z))

def relu(z): 
    return max(0, z)

x = np.array([1.0, 2.0])
w = np.array([0.5, -0.6])
b = 0.1

z = np.dot(x, w) + b
print("Linear Combination (z):", z)
print("Sigmoid Output:", sigmoid(z))
print("ReLU Output:", relu(z))
```
