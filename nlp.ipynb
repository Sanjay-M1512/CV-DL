```python
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http/S+|www/S+|https/S+', '', text)
    text = re.sub(r"/S+@/S+", '', text)
    text = re.sub(r'[^a-z/s]', '', text)
    text = re.sub(r'/d+', '', text)
    text = re.sub(r'/s+', ' ', text).strip()
    return text

input_text = "Hello123! Visit https://example.com or contact me at user@example.com. My age is 21!!!"
print("Preprocessed text:", preprocess_text(input_text))
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
text = '''Are you currently working on fine-tuning a BERT model or generating embeddings? 
Let me know what specific task you're trying to solve, 
and I can help with the code or data preparation! '''
tokens = tokenizer.tokenize(text)
print("Tokens:", tokens)

input_id = tokenizer(text, return_tensors="np")["input_ids"]
print("Input IDs:", input_id)

print(f"Input IDs to text: {tokenizer.decode(input_id[0])}")
# batch = ["Hello World", "BERT is great for NLP tasks", "Let's test tokenization"]
# b = tokenizer(batch, return_tensors="np")
# print(b)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

question = "Who directed the movie?"
context = "The movie was directed by X"

inputs = tokenizer(question, context, return_tensors="np")

input_ids = inputs["input_ids"]
print("Input IDs:", input_ids)
print(f"Decoded text: {tokenizer.decode(input_ids[0])}")
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

token = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

sentence = "BERT embeddings are useful for many NLP tasks."
inputs = token(sentence, padding=True, truncation=True,return_tensors="pt")

with torch.no_grad():
    outputs = model(**inputs)
    embeddings = outputs.last_hidden_state[:, 0, :]

display(embeddings)
```
