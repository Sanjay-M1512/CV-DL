```python
import numpy as np
import matplotlib.pyplot as plt

w0     = 10.0
steps  = 20
alphas = {"α=0.01 (too slow)": 0.01, "α=0.1 (appropriate)": 0.1, "α=1.1 (too large)": 1.1}
colors = ["steelblue", "seagreen", "tomato"]

plt.figure(figsize=(8, 4))

for (label, alpha), color in zip(alphas.items(), colors):
    w = w0
    history = [w]
    for _ in range(steps):
        w = w - alpha * 2 * w       
        history.append(w)
    loss = w ** 2
    print(f"{label:25s} → After {steps} steps: w = {w:>10.3f}, Loss = {loss:.3f}")
    plt.plot(history, label=label, color=color)

plt.axhline(0, color='black', linewidth=0.6, linestyle='--')
plt.title('Gradient Descent: w vs Iteration for Different Learning Rates')
plt.xlabel('Iteration')
plt.ylabel('w')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
