```python
import numpy as np
import matplotlib.pyplot as plt

w = 0.0
alpha = 0.1
steps = 10

history_w = []
history_loss = []

# Task 1: Gradient Descent Steps
for i in range(steps + 1):
    loss = (w - 3) ** 2
    history_w.append(w)
    history_loss.append(loss)
    if i in [0, 1, 5, 10]:
        print(f"Step {i:<2}: w = {w:.3f}, Loss = {loss:.3f}")
    if i < steps:
        grad = 2 * (w - 3)      # dL/dw
        w = w - alpha * grad    # update rule
```

```python
plt.figure(figsize=(7, 4))
plt.plot(range(steps + 1), history_loss, marker='o', color='steelblue')
plt.title('Gradient Descent Convergence: L(w) = (w − 3)²')
plt.xlabel('Iteration')
plt.ylabel('Loss L(w)')
plt.xticks(range(steps + 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```
