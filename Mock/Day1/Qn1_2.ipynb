```python
import numpy as np
import matplotlib.pyplot as plt

Z_values = [-2.0, -0.5, 0.0, 1.0, 3.5]

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def relu(z):
    return max(0, z)

print(f"{'Z':>6}  {'Sigmoid':>10}  {'ReLU':>6}")
print("-" * 28)
for z in Z_values:
    print(f"Z = {z:>4}  →  Sigmoid ≈ {sigmoid(z):.3f}  |  ReLU = {relu(z)}")

zero_relu = [z for z in Z_values if relu(z) == 0]
print(f"\nReLU outputs zero for Z values: {zero_relu}")
print("Implication: Neurons with Z ≤ 0 output zero gradient → 'dying ReLU' problem; those neurons stop learning.")
```

```python
Z = np.linspace(-5, 5, 300)

plt.figure(figsize=(7, 4))
plt.plot(Z, sigmoid(Z),       label='Sigmoid', color='steelblue')
plt.plot(Z, np.maximum(0, Z), label='ReLU',    color='tomato')
plt.axhline(0, color='black', linewidth=0.6, linestyle='--')
plt.axvline(0, color='black', linewidth=0.6, linestyle='--')
plt.title('ReLU vs Sigmoid Activation Functions')
plt.xlabel('Z (Input)')
plt.ylabel('Activation Output')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
